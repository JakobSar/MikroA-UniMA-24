## Aufgabe 1

```julia
using Plots
using LaTeXStrings

#Budgetreade
x1 = range(0, stop=5, length=400)
x2 = (3 .- 1.25x1) 
x5 = (4 .- 1.25x1)  
x6 = (5 .- 1.25x1)  
x7 = (2 .- 1.25x1)  
x8 = (1 .- 1.25x1) 
x9 = (6 .- 1.25x1)  

plot(x1, x2, label="Budgetgerade", xlabel=L"x_1", ylabel=L"x_2", title="Quasilineare Nutzenfunktion", linewidth=2, linecolor=:orange)
plot!(x1, x5, label="", linewidth=2, linecolor=:orange)
plot!(x1, x6, label="", linewidth=2, linecolor=:orange)
plot!(x1, x7, label="", linewidth=2, linecolor=:orange)
plot!(x1, x8, label="", linewidth=2, linecolor=:orange)
plot!(x1, x9, label="", linewidth=2, linecolor=:orange)

plot!(ylim=(0, 5), xlim=(0, 5))


for i in 0:0.1:10
    scatter!([0.8], [i], label="", markercolor=:purple, markersize=1)
end

#Indifferenzkurve
for i in 0:0.5:10
    x3 = range(0, stop=5, length=400)
    x4 = i .- sqrt.(5 .* x3)  # Nutzenfunktion
    if i == 0
        plot!(x3, x4, linewidth=2, linecolor=:green, label="Indifferenzkurven (Fixer Nutzen)")
    else
        plot!(x3, x4, linewidth=2, linecolor=:green, label="")
    end
end
scatter!([0.8], [2], color=:magenta, label="Optimale Allokation")
scatter!([0.8], [3], color=:magenta, label="")
scatter!([0.8], [4], color=:magenta, label="")
scatter!([0.8], [1], color=:magenta, label="")
scatter!([0.8], [0], color=:magenta, label="")
scatter!([0.8], [5], color=:magenta, label="")
# Show the plot
plot!()  # This line is necessary to display the plot
```

```julia
using Plots
using LaTeXStrings

# Werte fr x1 und x2
x1 = collect(0:0.1:15)  
x2 = collect(0:0.1:15)  

# Definiere die Leontief-Funktion
function leontief(x1, x2)
    return min(x1, x2)
end

output = [leontief(a, b) for a in x1, b in x2]

# Plotten der Leontief-Funktion als Konturplot
plot(x1, x2, output, xlabel=L"x_1", ylabel=L"x_2", title="Leontief-Nutzenfunktion")

plot!(ylim=(0, 15),xlim=(0,15))

x5 = range(0, stop=12, length=400)
x6 = x5
plot!(x5, x6, label="Preis-Konsum-Kurve", linewidth=2)

x3 = range(0, stop=15, length=400)
x4 = (12 .- 3x3) #p1=p2=1
x7 = (12 .- 2x3) #p1=p2=1
x8 = (9 .- 2x3) #p1=p2=1
plot!(x3, x4, label="Budgetgerade (alt Y,p)", xlabel=L"x_1", ylabel=L"x_2", linecolor=:green, linewidth=2)
plot!(x3, x7, linecolor=:darkgreen, linewidth=2, label="Budgetgerade (alt Y, neu p)")
plot!(x3, x8,linecolor=:lightgreen, linewidth=2, label="Budgetgerade (neu Y, neu p)")
scatter!([3], [3], color=:orange, label="Optimale Allokation")
```
