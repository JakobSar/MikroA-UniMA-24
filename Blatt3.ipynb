## Aufgabe 1

```julia
using Plots
using LaTeXStrings

x1 = range(0, stop=1, length=400)
x2 = 0.5(1 .- x1)  #p1<p2


plot(x1, x2, label="Budgetgerade", xlabel=L"x_1", ylabel=L"x_2", title="Nutzenfunktion und Indifferenzkurve, Fall p1 < p2", linewidth=3)
plot!(ylim=(-0.01, 1),xlim=(0,1.01))

for i in 0:0.25:3
    x3 = range(0, stop=1, length=400)
    x4 =  (i .- x3)
    if i == 0
        plot!(x3, x4, linewidth=2, linecolor=:green, label="Indifferenzkurven (Fixer Nutzen)")
    else
        plot!(x3, x4, linewidth=2, linecolor=:green, label="")
    end
    if  0.5 <= i <= 1.25
        annotate!(0.35, -0.3 + i, text("Nutzen = $i", :green, :left, :8, rotation=-30))  
    end
end
scatter!([1], [0], color=:orange, label="Optimale Allokation")
plot!() 
```

```julia
using Plots
using LaTeXStrings

x1 = range(0, stop=2, length=400)
x2 = 1.5(1 .- x1)  #p1>p2 

# Create the plot
plot(x1, x2, label="Budgetgerade", xlabel=L"x_1", ylabel=L"x_2", title="Nutzenfunktion und Indifferenzkurve, Fall p1 > p2", linewidth=3)
plot!(ylim=(0, 2),xlim=(-0.005,1))

for i in 0:0.25:3
    x3 = range(0, stop=1, length=400)
    x4 =  (i .- x3)
    if i == 0
        plot!(x3, x4, linewidth=2, linecolor=:green, label="Indifferenzkurven (Fixer Nutzen)")
    else
        plot!(x3, x4, linewidth=2, linecolor=:green, label="")
    end
    if  1 <= i <= 2
        annotate!(0.35, -0.25 + i, text("Nutzen = $i", :green, :left, :8, rotation=-15))  
    end
end
scatter!([0], [1.5], color=:orange, label="Optimale Allokation")
# Show the plot
plot!()  # This line is necessary to display the plot
```

## Aufgabe 2

```julia
using Plots
using LaTeXStrings

# Werte für x1 und x2
x1 = collect(0:0.01:10)  
x2 = collect(0:0.01:10)  

# Definiere die Leontief-Funktion
function leontief(x1, x2)
    return min(x1, x2)
end

output = [leontief(a, b) for a in x1, b in x2]

# Plotten der Leontief-Funktion als Konturplot
plot(x1, x2, output, xlabel="x_1", ylabel="x_2", title="Leontief-Nutzenfunktion")

plot!(ylim=(0, 10),xlim=(0,10))

x3 = range(0, stop=10, length=400)
x4 = (10 .- 1.65x3) #p1=p2=1
plot!(x3, x4, label="Budgetgerade", xlabel=L"x_1", ylabel=L"x_2", linewidth=2, linecolor=:green)
scatter!([5], [5], color=:orange, label="Optimale Allokation")
```

```julia
using Plots
using LaTeXStrings

# Create the plot
plot(ylim=(0, 1.1), xlim=(0,1.4), xlabel=L"k", ylabel=L"l", title="Indifferenzkurven \$(\\bar{l} = 0.7)\$")

for i in 0:0.13:5
    x3 = range(0, stop=5, length=400)
    x4 =  i .+ (x3 .- 0.7).^2 
    if i == 0
        plot!(x4, x3, linewidth=2, linecolor=:green, label="Indifferenzkurven (Fixer Nutzen)")
    else
        plot!(x4, x3, linewidth=2, linecolor=:green, label="")
    end
end

# Annotate the y-axis with the vertex point
annotate!([(-0.065, 0.7, text("\$\\bar{l}\$  -", 10,:left))])
annotate!([(1, -0.1, text("\$ k' \$", 10,:center))])
annotate!([(0.1, -0.1, text("\$ k_0 \$", 10,:center))])

# Budgetgerade
x1 = range(0, stop=2, length=400)
x2 = 0.5(1 .- x1)  # p1 > p2 
x5 = 0.5(1 .- 2.5x1)  # p1 > p2 
plot!(x1, x2, label="Budgetgerade A",  linewidth=2)
plot!(x1, x5, label="Budgetgerade A",  linewidth=2, legend=:topright)

# Show the plot
plot!()  # This line is necessary to display the plot
```
